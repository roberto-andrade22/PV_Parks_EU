## México

Mientras que Alemania en su mejor punto se reciben hasta 3.2 KWh por cada metro cuadrado al día de irradiación, en México en su peor punto recibe desde 3.6 KWh por cada metro cuadrado al día.

- Considerando el funcionamiento actual del Sistema Eléctrico Nacional, se podría estimar que la venta de electricidad por renovables sería a través de las subastas de mediano y largo plazo, y por venta de Certificados de energías limpias (CELs).
- Con base en estos supuestos, y considerando como piso la producción que se tuvo en Alemania y como techo la que se tuvo en España, se estaría hablando de ventas por entre 75 y 100 millones de USD cada año.

### Costos

En cuanto a los costos, la International Renewable Energy Agency (IRENA) ha reportado que los costos de producir energía solar se han desplomado un 82% en la última década, y se asume que éstos seguirán en su tendencia a la baja, a la par que se incrementa la eficiencia de las celdas a nivel comercial.

En economías emergentes, como India, los costos de capital ya se han logrado reducir a alrededor de 677 USD/kW a nivel *utility scale*.

Considerando las ventas mencionadas anteriormente, para instalar 1 GW de capacidad se estaría hablando de un periodo de retorno de inversión menor a 10 años. Sabiendo que los paneles producidos actualmente tienen una garantía y vida útil de 25 años y que los costos de operación son prácticamente nulos, se puede esperar una inversión redituable y confiable.
